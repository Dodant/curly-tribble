```python
import random
import matplotlib.pyplot as plt

filename = 'data/data-gp1.txt'

x_list = []
y_list = []

with open(filename, 'r') as f:
    for line in f:
        x, y = line.split(',')
        x_list.append(float(x))
        y_list.append(float(y))

plt.scatter(x_list, y_list)
```
