```python
import matplotlib.pyplot as plt
import math
import numpy

x = numpy.arange(-3, 3, 0.1)
y1 = []
y2 = []

for i in x:
    y1.append((1-math.exp(-i*2))/(1+math.exp(-i*2)))
    
for i in x:
    y2.append(1/(1+math.exp(-i)))
    
plt.axvline(x=0, color='black', linewidth=0.3)
plt.axhline(y=0, color='black', linewidth=0.3)
plt.axhline(y=1, color='blue', linewidth=0.3)
plt.axhline(y=-1, color='blue', linewidth=0.3)
plt.plot(x, y1, x, y2)
plt.show()
```
